## Figures for part 3: Varying the timescale separation between neural activity and HP

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Target Frequency = 0.1
OGfreqslow = np.loadtxt("OGfrequencySlowTarget.dat")
HPofffreqslow = np.loadtxt("HPofffrequencySlowTarget.dat")
```

```python
num_circuits = 50 #num circuits evolved per HPtau value
#num_repetitions = 
HPtaumin = .5
HPtaumax = 100
HPtaustep = .5
HPtauvals = np.arange(HPtaumin,HPtaumax+HPtaustep,HPtaustep)
num_HPtauvals = len(HPtauvals)

HPtauvalstile = np.repeat(HPtauvals,num_circuits)
HPtauvalstile = np.reshape(HPtauvalstile,(num_HPtauvals,num_circuits))

OGfreqslow = np.reshape(OGfreqslow,(num_HPtauvals,num_circuits))
HPofffreqslow = np.reshape(HPofffreqslow,(num_HPtauvals,num_circuits))
```

```python
plt.scatter(HPtauvalstile,OGfreqslow,s=3,label = 'Evolved Frequency')
plt.scatter(HPtauvalstile,HPofffreqslow,s=3,label = 'HP off')
plt.title('Osc. Freq. as Timescale of HP slows, n=50')
plt.xlabel(r'$\tau_{b} = \tau{w}$')
plt.ylabel("Osc. Frequency")
plt.legend()
```

```python
OGmean = np.mean(OGfreqslow,axis=1)
OGstd = np.std(OGfreqslow,axis=1)
HPoffmean = np.mean(HPofffreqslow,axis=1)
HPoffstd = np.std(HPofffreqslow,axis=1)

fig1,ax1 = plt.subplots()
ax1.errorbar(HPtauvals, OGmean,
            xerr=np.zeros(len(HPtauvals)),
            yerr=OGstd,
            fmt='-o',
            label = 'Evolved Frequency')

ax1.errorbar(HPtauvals, HPoffmean,
            xerr=np.zeros(len(HPtauvals)),
            yerr=HPoffstd,
            fmt='-o',
            label='HP off')

ax1.set_xlabel(r'$\tau_b=\tau_w$')
ax1.set_ylabel('Mean Osc. Freq. +/- 1sd')
ax1.set_title("Mean oscillatory frequency")

fig1.set_figwidth(50)
fig1.set_figheight(20)
plt.legend()
plt.show()
```

```python
#plot a subset of the data collected with a smaller range, less fine-grained

fig2,ax2 = plt.subplots()
ax2.errorbar(HPtauvals[1:-1:2], OGmean[1:-1:2],
            xerr=np.zeros(len(HPtauvals[1:-1:2])),
            yerr=OGstd[1:-1:2],
            fmt='-o',
            label = 'Evolved Frequency')

ax2.errorbar(HPtauvals[1:-1:2], HPoffmean[1:-1:2],
            xerr=np.zeros(len(HPtauvals[1:-1:2])),
            yerr=HPoffstd[1:-1:2],
            fmt='-o',
            label='HP off')

ax2.set_xlabel(r'$\tau_b=\tau_w$')
ax2.set_ylabel('Mean Osc. Freq. +/- 1sd')
ax2.set_title("Mean oscillatory frequency")

fig2.set_figwidth(10)
# fig2.set_figheight(20)

plt.show()
```
