```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CCdist = np.loadtxt('centercrossingdistance.dat')
CCdist = np.reshape(CCdist,(100000,3,2))
```

```python
avg_over_repeitions = np.average(CCdist,axis=1)
avg_over_repeitions_init = avg_over_repeitions[:,0]
avg_over_repeitions_final = avg_over_repeitions[:,1]
avg_over_repeitions_difference = avg_over_repeitions_final - avg_over_repeitions_init
```

```python
plt.boxplot(avg_over_repeitions_difference)
plt.title('Final CC-distance - Initial CC-distance')
```

```python
OGdim = np.loadtxt('OGdim.dat')
HPondim = np.loadtxt('HPondimension.dat')
HPoffdim = np.loadtxt('HPoffdimension.dat')
print(len(OGdim))
 
circuits_tested = (len(OGdim)-len(OGdim)%3)/3
print(circuits_tested)
sizes_fully_tested = int((circuits_tested-circuits_tested%100000)/100000)
print(sizes_fully_tested)

OGdim = OGdim[0:sizes_fully_tested*100000*3]
OGdim = np.reshape(OGdim, (sizes_fully_tested,100000,3))
OGdim = OGdim[:,:,-1] #accidentally kept track of the highest yet dimension after every repetition, but really only needed the end
HPondim = HPondim[0:sizes_fully_tested*100000*3]
HPondim = np.reshape(HPondim, (sizes_fully_tested,100000,3))
HPondim = HPondim[:,:,-1]
HPoffdim = HPoffdim[0:sizes_fully_tested*100000*3]
HPoffdim = np.reshape(HPoffdim, (sizes_fully_tested,100000,3))
HPoffdim = HPoffdim[:,:,-1]
```

```python
osc_or_not_OG = OGdim > 0
osc_or_not_HPon = HPondim > 0
osc_or_not_HPoff = HPoffdim > 0
# print(osc_or_not_HPoff[0:10])

prop_osc_OG = np.average(osc_or_not_OG,axis=1)
prop_osc_HPon = np.average(osc_or_not_HPon,axis=1)
prop_osc_HPoff = np.average(osc_or_not_HPoff,axis=1)
```

```python
sizes = np.linspace(1,sizes_fully_tested,sizes_fully_tested)
plt.plot(sizes,prop_osc_OG,label="init")
plt.plot(sizes,prop_osc_HPon,label="HPon")
plt.plot(sizes,prop_osc_HPoff,label="HPoff")
plt.title('Proportion oscillating')
plt.legend()
plt.xlabel('N')
plt.show()
```
