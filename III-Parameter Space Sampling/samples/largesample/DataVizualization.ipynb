```python
import numpy as np
import matplotlib.pyplot as plt
```

Percentage of circuits that oscillate before, during, and after HP across sizes
Sample size = 10,000

```python
OGdata = np.loadtxt('outputfileOGlargesamp.dat')
HPondata = np.loadtxt('outputfileHPonlargesamp.dat')
HPoffdata = np.loadtxt('outputfileHPofflargesamp.dat')
n = 10000
```

```python
OGdata = np.reshape(OGdata,(20,3))
OGpercentosc = OGdata[:,1]
OGpercICdep = (OGdata[:,2]/n)*100
HPondata = np.reshape(HPondata,(20,3))
HPonpercentosc = HPondata[:,1]
HPonpercICdep = (HPondata[:,2]/n)*100
HPoffdata = np.reshape(HPoffdata,(20,3))
HPoffpercentosc = HPoffdata[:,1]
HPoffpercICdep = (HPoffdata[:,2]/n)*100
```

```python
fig, ax = plt.subplots()
ax.plot(OGdata[:,0],OGpercentosc,label="no HP")
ax.fill_between(OGdata[:,0],OGpercICdep,0,color='blue',alpha=.1,label="no HP IC dep")
ax.plot(HPondata[:,0],HPonpercentosc,label="HP on")
ax.fill_between(OGdata[:,0],HPonpercICdep,0,color='orange',alpha=.2,label="HP on IC dep")
ax.plot(HPoffdata[:,0],HPoffpercentosc,label="HP turned off")
ax.fill_between(OGdata[:,0],HPoffpercICdep,0,color='green',alpha=.1,label="HP off IC dep")
ax.set_xticks([1,5,10,15,20])
ax.set_xlabel('# Neurons')
ax.set_ylabel('% osc. circuits')
ax.legend()
plt.show()
```
